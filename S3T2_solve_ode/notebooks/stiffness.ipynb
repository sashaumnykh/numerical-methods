```python
import numpy as np
from scipy.integrate import solve_ivp
# %matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
```

# Осциллятор Ван дер Поля
https://ru.wikipedia.org/wiki/Осциллятор_Ван_дер_Поля

* `x' = y`
* `y' = mu(1-x^2)y - x`

mu - параметр нелинейности затухания.

Этот пример показывает, как от параметра mu зависит то, какие методы будут плохо работать:
* mu = 0 - неявные
* mu >> 1 - явные

Возьмём и прогоним все методы, реализованные в SciPy, на этой задаче.

```python
def vdp(t, y, mu):
    return [
        y[1],
        mu*(1 - y[0]**2) * y[1] - y[0],
    ]

y0 = [2, 0]
t_span = [0, 2000]
methods = [
    'RK45', 
    'RK23', 
    'DOP853', 
    'Radau', 
    'BDF', 
    'LSODA'
]
```

```python
def test(mu, methods):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
    for method in methods:
        res = solve_ivp(vdp, y0=y0, t_span=t_span, method=method, args=[mu])
        print(f'{method:8}: {res.nfev:8} fevals, {len(res.t)-1:8} steps. {res.message}')
        ax1.plot(res.y[0], res.y[1], '.', label=method)
        ax1.set_title('y(x)')
        ax2.plot(res.t, res.y[0], '.:', label=method)
        ax2.set_title('x(t)')

    plt.legend()
    plt.suptitle(f'Van der Pol, mu={mu}')
    pass
```

### mu = 0
При mu=0 это простейшая колебательная система.

```python
mu = 0
test(mu, methods)
```

### mu = 10

```python
mu = 10
test(mu, methods)
```

### mu = 100

```python
mu = 100
test(mu, methods)
```

### mu = 1000
явными даже пробовать не будем здесь

```python
mu = 1000
test(mu, methods[3:])
```

### Посмотрим на собственные числа линеаризованной системы

```python
mu = 1000
res = solve_ivp(vdp, y0=y0, t_span=t_span, method='BDF', args=[mu])
xs, ys = res.y
ts = res.t

l1s, l2s = [], []
for x, y in zip(xs, ys):
    l1, l2 = sorted(np.real(np.linalg.eigvals([
        [0,           1          ], 
        [-2*mu*x*y-1, mu*(1-x**2)]
    ])))
    l1s.append(l1)
    l2s.append(l2)

plt.figure(figsize=(10, 6))
plt.plot(ts, l1s, '.:')
plt.plot(ts, l2s, '.:')
plt.xlabel('t')
plt.title(f'Eigenvalues, mu={mu}')
pass
```
